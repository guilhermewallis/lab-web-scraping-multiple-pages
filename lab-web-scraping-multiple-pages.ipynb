### a. From Wikipedia: List of Billboard Hot 100 top-ten singles in 2023

## 2. Practice web scraping

### a. List of links from Wikipedia page - Python_(programming_language)

```python
r = requests.get('https://en.wikipedia.org/wiki/Python_(programming_language)')
r.status_code
```

```python
html = r.content
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
links
```

### b. Number of titles that have changed in the USC since last release

```python
r = requests.get('https://uscode.house.gov/download/download.shtml')
r.status_code
```

```python
html = r.content
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
changed_titles = soup.find_all('div',attrs={'class':"usctitlechanged"})
len(changed_titles)
```

### c. Top ten FBI's Most Wanted list

```python
r = requests.get('https://www.fbi.gov/wanted/topten')
r.status_code
```

```python
html = r.content
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('h3',attrs={'class':"title"})
top10_list = []
for name in names:
    top10_list.append(name.get_text(strip=True))
top10_list
```
